```python
"""
1.	Data Preprocessing:
•	Load the dataset and print a sample number of rows.
•	Find the number of rows with any missing values. Remove any row with a missing value.
•	Convert 'Gender', 'Smoker', and 'Region' into numerical values suitable for regression analysis (e.g., using one-hot encoding for 'region' and
      binary encoding for 'sex' and 'smoker').
•	Normalize the features using Min-Max scaling.
"""
```

```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
insurance_df = pd.read_csv("insurance.csv")
insurance_df_original = insurance_df
insurance_df.head()
```

```python
# number of nuls per each column
insurance_df.isnull().sum()
```

```python
# dropping nulls and reseting the index
insurance_df_clean = insurance_df.dropna(ignore_index=True)
insurance_df_clean.isnull().sum() # re-check the sum again 
```

```python
# define one hot encoder 
enc = OneHotEncoder() 
# encode the three columns in dataframe 
enc_data = pd.DataFrame(enc.fit_transform( insurance_df_clean[['Gender', 'Smoker','Region']]).toarray())
# join encoded results with existing dataframe 
New_df = insurance_df_clean.join(enc_data)
New_df.columns = ['Age', 'Gender', 'BMI', 'Children','Smoker','Region', 'Expenses','female','male','noSmoker','Smoker','northeast','northwest', 'southeast', 'southwest'] 
print(enc.categories_)
print(New_df.columns)
```

```python
results = []
data = New_df[['Age', 'female','male', 'BMI', 'Children','noSmoker','northeast','northwest', 'southeast', 'southwest', 'Expenses']].copy()
data.head()
```

```python
# take out the features in an array and take out the lables in another array
features_raw = data[data.columns[:-1]].values
lables = data[data.columns[-1]].values
```

```python
# normalizing features 
scaler = MinMaxScaler() 
scaler.fit(features_raw)
print("the max values per each column: ",scaler.data_max_) 
print("the min values per each column: ",scaler.data_min_)  
print("the max values in features before scaling : ",features_raw.max()) 
print("the min values in features before scaling : ",features_raw.min()) 
features = scaler.transform(features_raw) # transform the features to Max min scaling
print("the max values in features after scaling : ",features.max()) 
print("the min values in features after scaling : ",features.min()) 
```

```python
"""
2.	Splitting the Data:
•	Divide the data into “features” and “target” subsets.
•	Split the data into training and testing subsets (commonly a 70/30, 75/25, or 80/20 split).
"""
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, lables, test_size=0.2, random_state=42)
y_train2= y_train.reshape(y_train.shape[0],1)
y_test2= y_test.reshape(y_test.shape[0],1)

print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
print("y_train: ", y_train.shape)
print("y_train reshaped", y_train2.shape)
print("y_test: ", y_test.shape)
print("y_train reshaped", y_test2.shape)
```

```python
"""
3.	Gradient Descent Implementation:
•	Implement the gradient descent algorithm (without the Scikit-Learn library) to find the regression line. Initialize parameters randomly and update them iteratively to minimize the loss function. Record the loss value for each iteration:
"""
```

```python
X_b = np.c_[np.ones((X_train.shape[0], 1)), X_train]
m = X_train.shape[0]  # number of data points (1056)
n = X_train.shape[1]   # number of features (10)

alpha = 0.01                 # Learning rate
n_iterations = 10000   # Number of iterations
# randomly intialize the values 
W = np.random.randn(X_train.shape[1]+1,1)   # Weight matrix (11, 1)
loss = []                          # Loss value for each iteration
print("X_b: ", X_b.shape)
print("number of data points: ", m)
print("number of features: ", n)
print("Weight matrix: ", W.shape)

for iteration in range(n_iterations): 
    gradients = 1/m * X_b.T.dot(X_b.dot(W) - y_train2)  
    W = W - alpha * gradients 
    predictions = X_b.dot(W) 
    # recording loss value each itration
    loss.append(mean_squared_error(y_train2, predictions))

print("gradients shape: ", gradients.shape)
print("predictions shape: ", predictions.shape)
```

```python
"""
4.	Model Evaluation:
•	Predict the expenses for the testing dataset using the trained model.
•	Compute the Mean Absolute Error (MAE) and Mean Squared Error (MSE) of the predictions.
•	Plot a histogram of the error distribution.

"""
```

```python
# use testing data for preditons 
X_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
gradients = 1/m * X_b.T.dot(X_b.dot(W) - y_test) 
predictions = X_b.dot(W) 
pred = predictions.reshape(1,predictions.shape[0])
yp = y_test2.reshape(1,y_test2.shape[0])
loss_GD_MSE = mean_squared_error(y_test2, predictions)
loss_GD_MAE = mean_absolute_error(y_test2, predictions)
plt.hist(yp[0] - pred[0], color = 'blue', edgecolor = 'black', bins = 10)
 
print("X_b: ", X_b.shape) 
print("Weight matrix: ", W.shape)
print("gradients shape: ", gradients.shape)
print("predictions shape: ", predictions.shape)
print("loss GD MSE", loss_GD_MSE)
print("loss GD MAE", loss_GD_MAE)
results = []
results.append(['Gradient Descent Implementation','alpha = 0.01',loss_GD_MSE,loss_GD_MAE])
```

```python
"""
5.	Learning Rate Analysis:
•	Demonstrate the effect of varying the learning rate on the convergence of the gradient descent algorithm.

"""
```

```python
# Repeating the last work with different Alpha 
def GD_func(alpha):
    X_b = np.c_[np.ones((X_train.shape[0], 1)), X_train]
    m = X_train.shape[0]  # number of data points (1056)
    n = X_train.shape[1]   # number of features (10)
 
    n_iterations = 10000   # Number of iterations
    # randomly intialize the values 
    W = np.random.randn(X_train.shape[1]+1,1)   # Weight matrix (11, 1)
    loss = [] 
    for iteration in range(n_iterations): 
        gradients = 1/m * X_b.T.dot(X_b.dot(W) - y_train2)  
        W = W - alpha * gradients 
        predictions = X_b.dot(W)  
    return mean_squared_error(y_train2, predictions), mean_absolute_error(y_train2, predictions)
```

```python
listOfAlphas = [0.01,0.05,0.1,0.2]
for alpha in listOfAlphas: 
    MSE,MAE = GD_func(alpha)
    print("alpha:" , alpha, "; MSE: ", MSE, "; MAE:", MAE)
    results.append(['Gradient Descent Implementation','alpha = ' + str(alpha),loss_GD_MSE,loss_GD_MAE])
```

```python
"""
6.	Scikit-learn Implementation:
 	Repeat the regression using the linear_model.LinearRegression class from scikit-learn.
 	Compute MAE and MSE for comparison.
"""
 
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
print("intercept: ",lin_reg.intercept_)
print("coef: ",lin_reg.coef_) 
predictions = lin_reg.predict(X_test)
print("predictions shape: ", predictions.shape)

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
loss_SL_MSE = mean_squared_error(y_test, predictions)
loss_SL_MAE = mean_squared_error(y_test, predictions)
results.append(['SciketLearn LinearRegression','',loss_SL_MSE,loss_SL_MAE])
```

```python
"""
7.	Normal Equation Implementation:
•	Use the normal equation method to find the regression line directly.
•	Compare the MAE and MSE with previous methods.

"""
```

```python
X_b = np.c_[np.ones((X_train.shape[0], 1)), X_train]
m = X_train.shape[0]  # number of data points (1056)
n = X_train.shape[1]   # number of features (10)
  
W = np.linalg.inv( X_b.T.dot(X_b)).dot(X_b.T).dot(y_train)
predictions = X_b.dot(W)  
 
loss_Eq_MSE = mean_squared_error(y_train2, predictions)
loss_Eq_MAE = mean_absolute_error(y_train2, predictions) 
 
print("X_b: ", X_b.shape) 
print("Weight matrix: ", W.shape)
print("gradients shape: ", gradients.shape)
print("predictions shape: ", predictions.shape)
print("loss Eq MSE", loss_Eq_MSE)
print("loss Eq MAE", loss_Eq_MAE)
results.append(['Normal Equation Implementation','',loss_Eq_MSE,loss_Eq_MAE])
```
